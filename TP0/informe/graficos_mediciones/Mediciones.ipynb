```python
import wave
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

----
# $$Mediciones\ Facudo$$
----

```python
medCharFacu = pd.read_csv("Facundo\Caracteres.txt", sep='\t')
medLineFacu = pd.read_csv("Facundo\Lineas.txt", sep='\t')
medWordFacu = pd.read_csv("Facundo\Palabras.txt", sep='\t')
```

# $Mediciones\ Caracteres$

```python
medCharFacu=medCharFacu.sort_values('Cantidad')
medCharFacu
```

# $Mediciones\ Lineas$

```python
medLineFacu=medLineFacu.sort_values('Cantidad') 
medLineFacu
```

# $Mediciones\ Palabras$

```python
medWordFacu=medWordFacu.sort_values('Cantidad')
medWordFacu
```

```python
fig = plt.figure(figsize=[16, 16])
ax = fig.add_subplot(1,1,1)
#plt.style.use("fast")
#Aca establezco la subdivicion del grafico
ax.set_xticks(np.linspace(0, 4, 17), minor=True)
ax.set_yticks(np.linspace(0,2.5, 26), minor=True)

ax.tick_params(axis='both', labelsize=30)

#Parametros en los cuatro lados
plt.tick_params(axis='both', labelsize=20, labelleft=True, labelright=False, labeltop=False, labelbottom=True)


#1000*EJERCICIO_A_1_A['time'] es para poner la escala en mili segundos

#SEÑAL DE ENTRADA Vi
plt.plot(medCharFacu['Cantidad'], medCharFacu['Real'],linewidth=4.0,c='r',marker='o', markersize=10,label='$Medición\ de\ Caracteres$')
plt.plot(medWordFacu['Cantidad'], medWordFacu['Real'],linewidth=4.0,c='g',marker='o', markersize=10,label='$Medición\ de\ Palabras$')
plt.plot(medLineFacu['Cantidad'], medLineFacu['Real'],linewidth=4.0,c='b',marker='o', markersize=10,label='$Medición\ de\ Lineas$')



#EJE HORIZONTAL
ax.axhline(0,linestyle='-',linewidth=4,c='k')
ax.axvline(100,linestyle='-',linewidth=5,c='k')

#Limites del grafico
plt.xlim(100, 100000000)
plt.ylim(0, 2.5)

#Nombre de los ejes
plt.xlabel('$Cantidad$', fontsize=30)
plt.ylabel('$Tiempo[s]$', fontsize=30)

#Parametros del recuadro con nombre de las curvas y tamaño de la fuente
plt.rcParams['figure.facecolor'] = 'white'  # bluish dark grey
plt.rcParams['axes.facecolor'] = 'white' # bluish dark grey
plt.rcParams['savefig.facecolor'] = 'white'  # bluish dark grey

for param in ['axes.labelcolor', 'xtick.color', 'ytick.color']:
    plt.rcParams[param] = 'k'  # very light grey
ax.set_xscale('log')

plt.rcParams['text.color']='k'
plt.grid(which='both', linestyle='--',color='grey')
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"


legend = ax.legend(loc=1, shadow=True, fontsize=18)
legend.get_frame().set_facecolor('0.9')
plt.savefig('medicionesFacundo.svg',format="svg",bbox_inches='tight')
```
